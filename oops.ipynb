### Constructors
#def __init__(self): ---> default constructor

#def __init__(self,clr,brand): ---> parameterised constructor

```python
class Car:
    def __init__(self):
        pass
    
    def __init__(self,clr,brand):
        self.clr=clr
        self.brand=brand
        print("Adding car to database...")
    
   
c1=Car('blue','bmw')
print(c1.clr)
print(c1.brand)
c2=Car('green','mercedes')
print(c2.clr)
print(c2.brand)
```

```python
class Student:
    name='no name'# Class Attribute
    def __init__(self,name,year):
        self.name=name # Obj Attribute
        self.year=year
s1=Student("Anas",1)
print(s1.name)
```

### Methods: 


```python
class Student:
    name='no name'# Class Attribute
    def __init__(self,name,year):
        self.name=name # Obj Attribute
        self.year=year
    def Welcome(self):
        print(f"Welcome {self.name}")
    def DisplayYear(self):
        print(f"{self.name} is in {self.year} year")
s1=Student("Anas",'2nd')
s1.Welcome()
s1.DisplayYear()
```

```python
class Student:
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2  
        self.m3=m3
    def Avg(self):
        avg=(self.m1+self.m2+self.m3)/3
        return avg
s1=Student(99,84,85)
print(f"{s1.Avg():.2f}")
```

```python
# Static methods are methods tht dont require self 
# to call the static method u need to use the decorator @staticmethod
class static:
    def __init__(self):
        pass
    @staticmethod
    def hello():
        print('hello')
s=static()
s.hello()
```

### Abstraction
#hiding the implentation of the class the returning only the essential features to the user


```python
class Car:
    def __init__(self):
        self.acc=False
        self.brk=False
        self.clutch=False
    def Start(self):
        self.acc=False
        self.clutch=False
        print("Car has started") #user sees only the essential messege tht the car has started
bmw=Car()
bmw.Start()
```

```python
class Account:
    def __init__(self,balance,account):
        self.balance=balance
        self.account=account
    #debit
    def debit(self,amt):
        self.balance -= amt
        print(f"Rs.{amt} was debited from account no:{self.account}")
    
    #credit
    def credit(self,amt):
        self.balance += amt
        print(f"Rs.{amt} was credited to account no:{self.account}")
    
    def show_bal(self):
        print(f"Your current balance is {self.balance}")
       
a1=Account(10000,123)
a1.debit(2000)
a1.show_bal()
a1.credit(4530)
a1.show_bal()
        
```
