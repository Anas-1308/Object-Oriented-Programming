### Inheritance

```python
class Car:
    color="black"
    @staticmethod
    def Start():
        print("car started...")
    @staticmethod
    def Stop():
        print("car stopped...")

class ToyotaCar(Car):
    def __init__(self,name):
        self.name=name
Car1=ToyotaCar("Camry")
Car1.Start()
Car1.color
```

```python
#Multi-level Inheritence
class Car:
    @staticmethod
    def Start():
        print("car started...")
    @staticmethod
    def Stop():
        print("car stopped...")

class ToyotaCar(Car):
    def __init__(self,brand):
        self.brand=brand

class Camry(ToyotaCar):
    def __init__(self, type):
        self.type=type

Car1=Camry("petrol")
Car1.Start()
```

```python
#super() method
class Car:
    def __init__(self,type):
        self.type=type
    @staticmethod
    def Start():
        print("car started...")
    @staticmethod
    def Stop():
        print("car stopped...")

class ToyotaCar(Car):
    def __init__(self,brand,type):
        super().__init__(type)   
        self.brand=brand
        
Car1=ToyotaCar("Camry","petrol")
Car1.type
```

### Polymorphism : Operator Overloading

```python
class Complex:
    def __init__(self,real,img):
        self.real=real
        self.img=img
    def showNumber(self):
        print(self.real,"i + ",self.img,"j")
    def __add__(self,num2):
        newReal = self.real + num2.real
        newimg=self.img + num2.img 
        return Complex(newReal,newimg)
    def __sub__(self,num2):
        newReal = self.real - num2.real
        newimg=self.img - num2.img 
        return Complex(newReal,newimg)
num1=Complex(2,6)
num1.showNumber()
num2=Complex(1,0)
num2.showNumber()

num3 = num1 + num2
num3.showNumber()
num4 = num1 - num2
num4.showNumber()

```

```python
class Circle:
    def __init__(self,r):
        self.r=r
    def Area(self):
        ar=3.14*(self.r)**2
        return ar
    def Perimeter(self):
        p=2*3.14*(self.r)
        return p
c1=Circle(3)
print(c1.Area())
print(c1.Perimeter())
```
